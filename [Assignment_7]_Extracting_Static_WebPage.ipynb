```python
import dateparser
```

To convert from thai text date string, we will use the parse method.  Note that the parse method assumes the BC year, not BE.  Thus, we will have to subtract 543 from the year.  In addition, weekday() returns day of week with 0=Monday, ..., 6=Sunday.

```python
dt = dateparser.parse('วันศุกร์ที่ 17 มกราคม 2563')
print(dt)
print(dt.weekday())
dt = dt.replace(year=dt.year-543)
print(dt)
print(dt.weekday())
```

## Assignment

Count the distribution of number of week days that are “วันพระ” for all three years and answer the following questions:

```python
import requests
from bs4 import BeautifulSoup
from bs4.element import Tag
from IPython.core.display import HTML

url1 = "https://www.myhora.com/%E0%B8%9B%E0%B8%8F%E0%B8%B4%E0%B8%97%E0%B8%B4%E0%B8%99/%E0%B8%A7%E0%B8%B1%E0%B8%99%E0%B8%9E%E0%B8%A3%E0%B8%B0-%E0%B8%9E.%E0%B8%A8.2563.aspx"
page1 = requests.get(url1)
html1 = page1.text

url2 = "https://www.myhora.com/%E0%B8%9B%E0%B8%8F%E0%B8%B4%E0%B8%97%E0%B8%B4%E0%B8%99/%E0%B8%A7%E0%B8%B1%E0%B8%99%E0%B8%9E%E0%B8%A3%E0%B8%B0-%E0%B8%9E.%E0%B8%A8.2564.aspx"
page2 = requests.get(url2)
html2 = page2.text

url3 = "https://www.myhora.com/%E0%B8%9B%E0%B8%8F%E0%B8%B4%E0%B8%97%E0%B8%B4%E0%B8%99/%E0%B8%A7%E0%B8%B1%E0%B8%99%E0%B8%9E%E0%B8%A3%E0%B8%B0-%E0%B8%9E.%E0%B8%A8.2565.aspx"
page3 = requests.get(url3)
html3 = page3.text
```

```python
soup1 = BeautifulSoup(html1, "html")
soup2 = BeautifulSoup(html2, "html")
soup3 = BeautifulSoup(html3, "html")
```

```python
all_date = []
all_date.append(soup1.select('div .bud-day .bud-day-col'))
all_date.append(soup2.select('div .bud-day .bud-day-col'))
all_date.append(soup3.select('div .bud-day .bud-day-col'))
all_date
```

```python
import pandas as pd

data = []

for y in range(3):
    idx = 0
    for e in all_date[y]:
        if idx % 3 == 0:
            data.append({ 'day' : e.string.strip()})
        idx += 1
    
    
df = pd.DataFrame(data)
df
```

```python
def thDateToWeekday(thDate):
  dt = dateparser.parse(thDate)
  dt = dt.replace(year=dt.year-543)
  return ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"][dt.weekday()]

df.day = df.day.apply(thDateToWeekday)
df
```

```python
df = df.day.value_counts()
df
```
